```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data from CSV files
hadoop_data = pd.read_csv('MapReduce_time/hive_weather_delay_time.csv')
spark_data = pd.read_csv('spark_time/spark_weather_delay_time.csv')
```

```python
# Ensure the data is sorted by iteration
hadoop_data.sort_values('Iteration', inplace=True)
spark_data.sort_values('Iteration', inplace=True)
```

```python
barWidth = 0.35

# Set position of bar on X axis
r1 = np.arange(len(hadoop_data['Iteration']))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, hadoop_data['Execution'], color='b', width=barWidth, edgecolor='grey', label='Hadoop')
plt.bar(r2, spark_data['ExecutionTime'], color='r', width=barWidth, edgecolor='grey', label='Spark')

# Add xticks on the middle of the group bars
plt.xlabel('Iteration', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(hadoop_data['Iteration']))], hadoop_data['Iteration'])

# Create legend & Show graphic
plt.ylabel('Running Time')
plt.title('Running Time vs Iteration for Hadoop and Spark - Weather delay')
plt.legend()
plt.show()
```

```python
# Calculate the average running time
hadoop_avg_time = hadoop_data['Execution'].mean()
spark_avg_time = spark_data['ExecutionTime'].mean()

# Print the average running times
print(f"Hadoop Average Running Time: {hadoop_avg_time} units")
print(f"Spark Average Running Time: {spark_avg_time} units")
```
